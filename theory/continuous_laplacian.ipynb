```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def f1(x):
    return x

x1 = np.linspace(-5, 5, 100)
y1 = f1(x1)

def f2(x):
    return x**2

x2 = np.linspace(-5, 5, 100)
y2 = f2(x2)

def f3(x):
    return np.sin(x)/x

x3 = np.linspace(-15, 15, 300)
y3 = f3(x3)

def f4(x):
    return -np.sin(x)/x - 2.0*np.cos(x)/x**2 + 2.0*np.sin(x)/x**3

y4 = f4(x3)
```

```python
fig = pl.figure(figsize=(14, 14), layout="constrained")
gs = fig.add_gridspec(2, 4)

ax00 = fig.add_subplot(gs[0, 0:2])
ax00.plot(x1, y1, color="black", linewidth=2, label=r"$f_1(x)=x$")
ax00.grid(True)
ax00.axvline(0, color="red", linestyle="--")
ax00.axhline(0, color="red", linestyle="--")
ax00.vlines([2.0, 3.0, 4.0], [0.0, 0.0, 0.0], [f1(2.0), f1(3.0), f1(4.0)], color="green", linestyle="--")
ax00.plot([2.0, 3.0, 4.0], [f1(2.0), f1(3.0), f1(4.0)], color="green", linestyle="", marker='o')
ax00.text(2.0, -0.8, r'$a-\varepsilon$', rotation=45, fontsize=10, color='green', ha='center', va='bottom')
ax00.text(3.0, -0.6, r'$a$', rotation=45, fontsize=10, color='green', ha='center', va='bottom')
ax00.text(4.0, -0.8, r'$a+\varepsilon$', rotation=45, fontsize=10, color='green', ha='center', va='bottom')
section = np.linspace(2.0, 4.0, 10)
ax00.fill_between(section, f1(section), alpha=0.1, color="green", label=r"$2\varepsilon\bar{f}_1(a)$")
ax00.fill_between(section, f1(3.0)*np.ones_like(section), alpha=0.1, color="orange", edgecolor="darkorange", linewidth=2, label=r"$2\varepsilon{f}_1(a)$")
ax00.legend(loc="upper left")

ax01 = fig.add_subplot(gs[0, 2:])
ax01.plot(x2, y2, color="black", linewidth=2, label=r"$f_2(x)=x^2$")
ax01.grid(True)
ax01.axvline(0, color="red", linestyle="--")
ax01.axhline(0, color="red", linestyle="--")
ax01.vlines([2.0, 3.0, 4.0], [0.0, 0.0, 0.0], [f2(2.0), f2(3.0), f2(4.0)], color="green", linestyle="--")
ax01.plot([2.0, 3.0, 4.0], [f2(2.0), f2(3.0), f2(4.0)], color="green", linestyle="", marker='o')
ax01.set_ylim(-2.0, 20.0)
ax01.text(2.0, -1.5, r'$a-\varepsilon$', rotation=45, fontsize=10, color='green', ha='center', va='bottom')
ax01.text(3.0, -1.2, r'$a$', rotation=45, fontsize=10, color='green', ha='center', va='bottom')
ax01.text(4.0, -1.5, r'$a+\varepsilon$', rotation=45, fontsize=10, color='green', ha='center', va='bottom')
section = np.linspace(2.0, 4.0, 10)
ax01.fill_between(section, f2(section), alpha=0.1, color="green", label=r"$2\varepsilon\bar{f}_2(a)$")
ax01.fill_between(section, f2(3.0)*np.ones_like(section), alpha=0.1, color="orange", edgecolor="darkorange", linewidth=2, label=r"$2\varepsilon{f}_2(a)$")
ax01.legend(loc="upper center")

ax10 = fig.add_subplot(gs[1, 1:3])
ax10.plot(x3, y3, color="black", linewidth=2, label=r"$f_3(x)=\frac{sin(x)}{x}$")
ax10.plot(x3, y4, color="blue", linewidth=0.5, label=r"$\Delta f_3(x)$")
ax10.grid(True)
ax10.axvline(0, color="red", linestyle="--")
ax10.axhline(0, color="red", linestyle="--")
ax10.fill_between(x3, 0.0, y4, where=(y4 <= 0.0), alpha=0.1, color="red", label=r"$f_3(x)\geq\bar{f}_3(x)$")
ax10.fill_between(x3, 0.0, y4, where=(y4 >= 0.0), alpha=0.1, color="purple", label=r"$f_3(x)\leq\bar{f}_3(x)$")
ax10.legend(loc="upper right")

pl.savefig("./images/continuous_laplacian.pdf", bbox_inches='tight')
```
